```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from nltk.stem import WordNetLemmatizer
import re

import nltk
nltk.download('wordnet')
!unzip /usr/share/nltk_data/corpora/wordnet.zip -d /usr/share/nltk_data/corpora/

stopwords.words('english')

# 初始化必要的資源
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def word_preprocessor(text):
    # 轉為小寫
    text = text.lower()  # 小寫
    text = re.sub(r'\d+', '', text)  # 移除數字
    text = re.sub(r'\W+', ' ', text)  # 移除符號
    text = re.sub(r'\s+', ' ', text)
    words = word_tokenize(text)
    processed_words = [
        lemmatizer.lemmatize(word) for word in words
        if word not in stop_words and len(word) > 2  # 去除停用詞和過短的詞
    ]
    return processed_words
```

```python
from gensim.models.ldamodel import LdaModel
num_topics = 10
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
for idx, topic in lda_model.show_topics(num_topics=num_topics, formatted=False):
    print(f"Topic {idx}: {[word for word, _ in topic]}")
```

```python
def get_dominant_topic(bow):
    """
    輸入文檔的 Bag-of-Words (BoW)，返回主題 ID 和對應的概率。
    """
    topic_probs = lda_model.get_document_topics(bow)
    dominant_topic = max(topic_probs, key=lambda x: x[1])  # 找到概率最高的主題
    return dominant_topic[0], dominant_topic[1]  # 返回主題 ID 和概率

# 為每篇文檔推斷前三個主題
def get_top_topics(bow, top_n=3):
    """
    輸入文檔的 Bag-of-Words (BoW)，返回概率最大的前 top_n 個主題及其概率。
    """
    topic_probs = lda_model.get_document_topics(bow)
    sorted_topics = sorted(topic_probs, key=lambda x: x[1], reverse=True)[:top_n]  # 按概率排序
    top_topics = [topic[0] for topic in sorted_topics]  # 提取主題 ID
    top_probs = [topic[1] for topic in sorted_topics]  # 提取概率
    return top_topics, top_probs


# 應用到訓練數據
train_df['bow'] = processed_docs.apply(dictionary.doc2bow)  # 將每篇文檔轉為 BoW 格式
train_df['dominant_topic'], train_df['dominant_prob'] = zip(*train_df['bow'].map(get_dominant_topic))  # 新增主題欄位
```

```python
# 查看測試數據中每個主題組合在訓練數據中的數量
def count_topic_combinations(test_topics, train_dominant_topics):
    """
    計算測試數據中每個主題組合在訓練數據中的出現次數。
    """
    from collections import Counter
    train_topic_counts = Counter(train_dominant_topics)
    return test_topics.apply(lambda x: sum(train_topic_counts.get(topic, 0) for topic in x))

test_df['processed_prompt'] = test_df['prompt'].apply(word_preprocessor)
test_df_filtered = test_df[test_df['processed_prompt'].apply(len) > 4]
processed_docs_test = test_df['processed_prompt']
test_df['bow'] = processed_docs_test.apply(dictionary.doc2bow)  # 將每篇文檔轉為 BoW 格式
test_df['top_topics'], test_df['top_probs'] = zip(*test_df['bow'].map(lambda x: get_top_topics(x, top_n=3)))  # 新增主題欄位
test_df['train_topic_counts'] = count_topic_combinations(test_df['top_topics'], train_df['dominant_topic'])

# 查看測試結果
test_df[['prompt', 'top_topics', 'train_topic_counts']]
```

```python
import re

def is_code(text):
    """
    檢查是否包含程式碼特徵
    """
    code_patterns = [
        r'\b(def|class|import|return|else|while)\b',  # Python 關鍵字
        r'[(){}\[\];:]',  # 程式碼常見的括號和分隔符號
        r'\b(int|str|list|dict|float|bool)\b',  # Python 常見的資料型別
        r'[=+\-*/]',  # 運算符號
    ]
    
    # 計算匹配的程式碼特徵數量
    code_score = sum(1 for pattern in code_patterns if re.search(pattern, text))
    
    # 如果匹配到 2 個以上的特徵，則判定為程式碼
    return code_score >= 4

# 使用新的 is_code 函數過濾資料
def filter_data(df):
    """
    過濾資料：刪除程式碼樣本
    """
    mask = ~df['response_a'].apply(is_code) & ~df['response_b'].apply(is_code)
    return df[mask]

# no_code_train_df = filter_data(df_filtered)
no_code_train_df = filter_data(train_df)
no_code_train_df
```

```python
class PreferenceDataset(torch.utils.data.Dataset):
    def __init__(self, dataframe):
        self.data = dataframe

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return {
            "response_a": self.data.iloc[idx]["response_a"],
            "response_b": self.data.iloc[idx]["response_b"],
            "class_label": self.data.iloc[idx]["class_label"],
        }
```

```python
tokenizer = T.BertTokenizer.from_pretrained("/kaggle/input/google_bert_uncasedd/pytorch/default/1", cache_dir="./cache/")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def collate_fn(batch):
    pt_response_a = [item["response_a"] for item in batch]
    pt_response_b = [item["response_b"] for item in batch]
    class_label = [item["class_label"] for item in batch]
    
    # 分別對句子 A 和句子 B 進行編碼
    encoding_a = tokenizer(
        pt_response_a,
        padding=True,
        truncation=True,
        return_tensors="pt",
        max_length=256,  # 設定最大長度
    )
    encoding_b = tokenizer(
        pt_response_b,
        padding=True,
        truncation=True,
        return_tensors="pt",
        max_length=256,
    )
    
    # 返回兩個編碼的結果，並包括標籤
    return {
        "input_ids_a": encoding_a["input_ids"].to(device),
        "attention_mask_a": encoding_a["attention_mask"].to(device),
        "input_ids_b": encoding_b["input_ids"].to(device),
        "attention_mask_b": encoding_b["attention_mask"].to(device),
        "labels": torch.tensor(class_label).to(device),  # 標籤張量
    }

# 假設 PreferenceDataset 已正確實現
dataset_train = PreferenceDataset(no_code_train_df)

# dataloader
dl_train = DataLoader(dataset_train, batch_size=8, shuffle=True, collate_fn=collate_fn)
```

```python
import torch
import torch.nn as nn
from transformers import AutoModel

class MultiSentencePairModel(nn.Module):
    def __init__(self, pretrained_model="/kaggle/input/google_bert_uncasedd/pytorch/default/1"):
        super().__init__()
        self.bert = AutoModel.from_pretrained(pretrained_model)
        # self.dropout = nn.Dropout(0.2)
        self.classification_head = nn.Linear(768 * 2, 3)  # 拼接後維度是 768 * 2

    def forward(self, input_ids_a, attention_mask_a, input_ids_b, attention_mask_b):
        # 獲取句子 A 的嵌入
        outputs_a = self.bert(
            input_ids=input_ids_a, attention_mask=attention_mask_a
        )
        pool_output_a = outputs_a.pooler_output  # [CLS] token 的嵌入

        # 獲取句子 B 的嵌入
        outputs_b = self.bert(
            input_ids=input_ids_b, attention_mask=attention_mask_b
        )
        pool_output_b = outputs_b.pooler_output  # [CLS] token 的嵌入

        # 合併嵌入（拼接）
        combined_output = torch.cat([pool_output_a, pool_output_b], dim=1)

        # 添加 Dropout 和分類頭
        # combined_output = self.dropout(combined_output)
        logits = self.classification_head(combined_output)

        return logits

# 創建模型
model = MultiSentencePairModel().to(device)
```

```python
# optimizer
from torch.utils.data import DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AdamW
from tqdm import tqdm
optimizer = AdamW(model.parameters(), lr=3e-5)
classification_loss_fn = nn.CrossEntropyLoss()
```

```python
from tqdm import tqdm

epochs = 3

for epoch in range(epochs):
    model.train()
    bar = tqdm(dl_train)
    bar.set_description(f"Training epoch [{epoch+1}/{epochs}]")

    for batch in bar:
        optimizer.zero_grad()

        # 提取批次數據
        input_ids_a = batch["input_ids_a"]
        attention_mask_a = batch["attention_mask_a"]
        input_ids_b = batch["input_ids_b"]
        attention_mask_b = batch["attention_mask_b"]
        labels = batch["labels"]

        # 前向傳播
        outputs = model(
            input_ids_a=input_ids_a,
            attention_mask_a=attention_mask_a,
            input_ids_b=input_ids_b,
            attention_mask_b=attention_mask_b,
        )

        # 計算損失
        loss = classification_loss_fn(outputs, labels)

        # 反向傳播和優化
        loss.backward()
        optimizer.step()

        # 可選：更新學習率調度器
        # scheduler.step()

        # 更新進度條
        bar.set_description(f"Epoch {epoch+1}/{epochs} | Loss: {loss.item():.4f}")
```

```python
class TestPreferenceDataset(torch.utils.data.Dataset):
    def __init__(self, dataframe):
        self.data = dataframe

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return {
            "response_a": self.data.iloc[idx]["response_a"],
            "response_b": self.data.iloc[idx]["response_b"],
        }
```

```python
import transformers as T
import torch
from torch.utils.data import DataLoader

tokenizer = T.BertTokenizer.from_pretrained("/kaggle/input/google_bert_uncasedd/pytorch/default/1", cache_dir="./cache/")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def collate_fn_test(batch):
    pt_response_a = [item["response_a"] for item in batch]
    pt_response_b = [item["response_b"] for item in batch]
    
    # 分別對句子 A 和句子 B 進行編碼
    encoding_a = tokenizer(
        pt_response_a,
        padding=True,
        truncation=True,
        return_tensors="pt",
        max_length=256,  # 設定最大長度
    )
    encoding_b = tokenizer(
        pt_response_b,
        padding=True,
        truncation=True,
        return_tensors="pt",
        max_length=256,
    )
    
    # 返回兩個編碼的結果，並包括標籤
    return {
        "input_ids_a": encoding_a["input_ids"].to(device),
        "attention_mask_a": encoding_a["attention_mask"].to(device),
        "input_ids_b": encoding_b["input_ids"].to(device),
        "attention_mask_b": encoding_b["attention_mask"].to(device),
    }

# 假設 PreferenceDataset 已正確實現
dataset_test = TestPreferenceDataset(test_df)

# dataloader
dl_test = DataLoader(dataset_test, batch_size=8, shuffle=True, collate_fn=collate_fn_test)
```

```python
import torch.nn.functional as F

def predict(model, dataloader):
    model.eval()  # 設置模型為評估模式
    probabilities = []  # 用於儲存機率

    with torch.no_grad():  # 禁用梯度計算
        for batch in dataloader:
            input_ids_a = batch["input_ids_a"]
            attention_mask_a = batch["attention_mask_a"]
            input_ids_b = batch["input_ids_b"]
            attention_mask_b = batch["attention_mask_b"]

            # 前向傳播
            outputs = model(
            input_ids_a=input_ids_a,
            attention_mask_a=attention_mask_a,
            input_ids_b=input_ids_b,
            attention_mask_b=attention_mask_b,
            )

            # 應用 softmax 將 logits 轉為機率
            probs = F.softmax(outputs, dim=1)  # dim=1 表示沿每個類別的維度計算
            probabilities.extend(probs.cpu().numpy())  # 將結果轉為 numpy 並保存

    return probabilities  # 返回機率分佈

predictions = np.array(predict(model, dl_test), dtype=np.float64)
probabilities = pd.DataFrame(predictions, columns=['winner_model_a','winner_model_b','winner_tie'])
submission = test_df[["id"]].join(probabilities)
submission.to_csv('submission.csv', index=False)

print(submission.info())
submission
```
